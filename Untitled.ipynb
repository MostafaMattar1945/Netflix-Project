```python
#finding House of Cards show using isin() and str.contain()

data[data['Title'].isin(['House of Cards'])]
data[data['Title'].str.contains('House of Cards')]
```

```python
#showwing the year that has the most movies and TV shows
#first of all creating a new row that converts the object date to datetime date

data['NEW_R_Date'] = pd.to_datetime(data['Release_Date'], errors='coerce')
```

```python
# countingt how many each year appers in the dataset which makes us know which yeaqrv has the most number of movies and tv shows

data['NEW_R_Date'].dt.year.value_counts()
```

```python
#bar chart  version of the data

data['NEW_R_Date'].dt.year.value_counts().plot(kind='bar')
```

```python
#counting how many movies and tv shows in the dataset
data.groupby('Category').Category.count().plot(kind= 'bar')
```

```python
#showing all movies released in year 2020

data['year'] = data['NEW_R_Date'].dt.year  #creating year column
data[(data['Category'] == 'Movie') & (data['year']== 2020)] 
```

```python
#showing tv shows in india only
data[(data['Country']== 'India') & (data['Category'] == 'TV Show')]
```

```python
#max duration of movies or shows in netflix

data[['Mins', 'Unit']] = data['Duration'].str.split(' ', expand=True)
data['Mins'] = pd.to_numeric(data['Mins'], errors='coerce')
data.Mins.max()
data.Mins.mean()
print(f'the mean is {data.Mins.mean()} while the maximum is {data.Mins.max()}')
```
